```python
words = open("names.txt", 'r').read().splitlines()
```

```python
print("Total number of data: " + str(len(words)))

print("\nFirst 10 names: ")
print(words[:10])
```

```python
print("lengths of shortest name:", min(len(w) for w in words)," and longest name:", max(len(w) for w in words))
```

### Check bi-grams for first 3 names

```python
b = {} # initializing an empty dictionary to store the bigram counts
for w in words[:3]:
    print ("\nName:", w)
    characters = ['<S>']+list(w)+['<E>']
    for c1, c2 in zip(characters, characters[1:]):
        bigram = (c1, c2)
        b[bigram] = b.get(bigram, 0) + 1 # counting the bigrams in a dictionary, if not found return 0 and add 1 to it
        print(c1, c2)
```

```python
# count of bigrams in the dictionary
b
```

### Bi-gram counts for entire dataset

```python
b = {} # initializing an empty dictionary to store the bigram counts
for w in words:
    characters = ['<S>']+list(w)+['<E>']
    for c1, c2 in zip(characters, characters[1:]):
        bigram = (c1, c2)
        b[bigram] = b.get(bigram, 0) + 1 # counting the bigrams in a dictionary, if not found return 0 and add 1 to it
```

```python
sorted(b.items(), key=lambda item: item[1], reverse=True) # item[1] is the value and item[0] is the key
```

### PyTorch

```python
import torch
```

```python
N = torch.zeros((28, 28), dtype=torch.int32)
```

```python
characters = sorted(list(set(''.join(words))))
characters
```

```python
s_to_i = {s:i for i, s in enumerate(characters)} # mapping characters to integers

s_to_i['<S>'] = 26 # adding start token to the mapping
s_to_i['<E>'] = 27 # adding end token to the mapping

s_to_i
```

```python
for w in words:
    characters = ['<S>']+list(w)+['<E>']
    for c1, c2 in zip(characters, characters[1:]):
        ix1=s_to_i[c1] # getting the integer index of the first character
        ix2=s_to_i[c2] # getting the integer index of the second character
        N[ix1, ix2] += 1 # incrementing the count of the bigram in the matrix
```
